```python
import pandas as pd
```

```python
chicago = pd.read_csv('chicago.csv').dropna(how='all')
chicago.nunique()
```

```python
chicago['Department'] = chicago['Department'].astype('category')
```

```python
chicago.info()
```

```python
chicago.head(10)
```

```python
chicago['Name']=chicago['Name'].str.title()
chicago
```

```python
chicago['Department'] = chicago['Department'].str.replace('MGMNT', 'MANAGEMENT')
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '').astype(float)
chicago
```

```python
# filtering dataframe row with string methods

query = chicago['Position Title'].str.lower().str.contains('water')
chicago[query]
```

```python
mask = chicago['Position Title'].str.lower().str.startswith('water') #there's an endswith method
chicago[mask]
```

```python
chicago.columns = chicago.columns.str.upper()
chicago
```
