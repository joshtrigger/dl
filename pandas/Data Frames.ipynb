```python
import math

def convert_salary(salary):
    if math.isnan(salary):
        return '-'
    return f'${math.ceil(salary / 1000000)}M'
```

```python
df = pd.read_csv('employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.head()
```

```python
female_mask = df['Gender'] == 'Female'
df[female_mask]
```

```python
finance_mask = df['Team'] == 'Finance'
df[finance_mask]
```

```python
salary_mask = df['Salary'] > 110000
df[salary_mask]
```

```python
mask = female_mask & finance_mask & df['Senior Management'] & salary_mask
df[mask]
```

```python
df.info()
df[female_mask | finance_mask | salary_mask]
```

```python
mask1=df['First Name']=='Robert'
mask2=df['Team']=='Client Services'
mask3=df['Start Date']>'2016-06-01'
df[(mask1 & mask2 )| mask3]
```

```python
# isin method filters a single series with one or more values

df[df['Team'].isin(['Legal', 'Product', 'Sales'])]
```

```python
team_null_mask = df['Team'].isnull() & df['Gender'].isnull()
df2 = df[team_null_mask]
df2[df2['Team'].notnull()]
```

```python
df[df['Salary'].between(60000, 70000)].info()
```
