```python
import pandas as pd
```

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace=True)
```

```python
# works same as index_col param (can be more than one level)
# reset_index method undoes set_index
# set_index tosses col out of memory if used more than once

bond.set_index(['Film', 'Year']).head(3)
```

```python
# set a new value for the extracted cell
bond.loc['Spectre', ['Actor', 'Director']] = ['Joshua Lugada', 'John Doe']
bond.loc['Spectre', ['Actor', 'Director']]
```

```python
# overwrite multiple values

sean = bond['Actor'] == 'Sean Connery'
bond.loc[sean, 'Actor'] = 'Joshua Dan Lugada'
bond
```

```python
# rename() renames row and column names 

bond.rename(mapper = {'Skyfall':'Hello World'}, axis = 0)
bond.rename(mapper = {'Year':'Release','Box Office':'Revenue'}, axis = 1)
```

```python
# delete rows and columns using .drop()

bond.drop('Casino Royale')
bond.drop('Bond Actor Salary', axis = 1)
```

```python
def convert_to_string(number):
    return '$' + str(number)
```

```python
cols = ['Box_Office', 'Budget', 'Bond_Actor_Salary']
for col in cols:
    bond[col] = bond[col].apply(convert_to_string)
bond
```

```python
directors = bond['Director'].copy()
directors['A View to a Kill'] = 'Mr. Joshua Lugada'
```
